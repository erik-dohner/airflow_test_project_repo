```python
import pandas as pd
from api_library import BQConnect

# Init the BQConnect class
bq_connection = BQConnect()

# Create the query 
query = """
select
    *
from
    `dbt-test-449821.pipeline.int_sm_joined`
"""
```

```python
# Run the query 
df  = bq_connection.client.query(query).to_dataframe()
df
```

```python
pivoted_df = df.pivot(
    index='respondent_id',
    columns='question_text',
    values='answer_text'
)

# Reset the pivot tables index to return to regular tabular format
pivoted_df = pivoted_df.reset_index()
```
